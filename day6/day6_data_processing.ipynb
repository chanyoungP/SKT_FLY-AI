### 크롤링으로 워드클라우드

```python
from bs4 import BeautifulSoup
import requests
import time

user_agent = {"User-Agent": "Mozilla/5.0 (Macintosh;Intel Mac OS X 10_15_4) AppleWebKit/537.36 (KHTML,like Gecko) Chrome/83.0.4103.97 Safari/537.36"}
questions = []

URL = "https://qna.programmers.co.kr/?page={}"

for i in range(1,30):
  res = requests.get(URL.format(i),user_agent)
  soup = BeautifulSoup(res.text,'html.parser')
  ul_tags = soup.find_all("ul",'question-tags')
  for ul in ul_tags:
    li_tags = ul.find_all('li')
    for li in li_tags:
      questions.append(li.text.strip())
  time.sleep(0.5)
```

```python
text_file = ' '.join(questions)
text_file
```

```python
wordcloud = WordCloud().generate(text_file)
wordcloud.words_
```

```python
# 마스크를 이용한 시각화
from PIL import Image
import numpy as np

word_mask = np.array(Image.open('./grey_heart.png'))
word_mask.shape
# word_mask_1 = word_mask[:,:,0]
```

```python
plt.figure(figsize = (20,10))
STOPWORDS.add('python')
wordcloud = WordCloud(stopwords = STOPWORDS,max_words=50,mask = word_mask,background_color= 'ivory').generate(text_file)
plt.imshow(wordcloud)
plt.axis('off')
plt.show
```

```python


from bs4 import BeautifulSoup
import requests
import time
from wordcloud import WordCloud
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

user_agent = {"User-Agent": "Mozilla/5.0 (Macintosh;Intel Mac OS X 10_15_4) AppleWebKit/537.36 (KHTML,like Gecko) Chrome/83.0.4103.97 Safari/537.36"}
papers = []

URL = "https://neurips.cc/virtual/2023/papers.html?filter=titles"

res = requests.get(URL,user_agent)

soup = BeautifulSoup(res.text,'html.parser')

title = soup.select('li')
title_list = [i.text for i in title]
title_list

text_data = ' '.join(title_list)


# 마스크를 이용한 시각화
from PIL import Image
import numpy as np

word_mask = np.array(Image.open('./brain_mask.jpg'))

wordcloud = WordCloud(max_words = 100, mask = word_mask,
                      background_color= 'ivory').generate(text_data)
plt.figure(figsize = (10,10))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

```python
custom = set(['via','based']+['for','and','with','in','A','of','the','on','to','from'])

wordcloud = WordCloud(max_words = 20, mask = word_mask,stopwords = custom,
                      background_color= 'ivory').generate(text_data)

x,y = wordcloud.words_.keys(), wordcloud.words_.values()
plt.figure(figsize = (10,10))
plt.bar(x,y)
plt.xticks(rotation='vertical')  # x축 눈금 레이블을 세로로 표시
plt.show()
```
